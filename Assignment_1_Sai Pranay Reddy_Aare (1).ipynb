#### 17) Convert the Date column to the datetime type. Filter the data so it contains just the rows for 2015, for the conventional type, and for the region named Albany. Store just the Date, Total Bags, and Small Bags columns in a new DataFrame. (point: 10)

```python
avocado_data['Date'] = pd.to_datetime(avocado_data['Date'])
new_data = avocado_data.query("year == 2015 and type == 'conventional' and region == 'Albany'")
new_data=new_data.loc[:, ['Date', 'TotalBags', 'SmallBags']]
new_data
```

#### 18) Melt the data in the Total Bags and Small Bags columns, but not the values in the Date column. Name the column that contains the type of bag Bags, and name the column that contains the number of bags Count. Then, display the resulting DataFrame. (point: 10)

```python
melted_df = pd.melt(new_data, id_vars=["Date"], value_vars=["TotalBags", "SmallBags"], var_name="Bags", value_name="Count")
melted_df = melted_df.rename(columns={"value": "Count"})
melted_df
```

#### 19) Plot the melted data with Seaborn in a line plot, using the hue parameter to distinguish between the bag types. (point: 5)

```python
sns.lineplot(x="Date", y="Count", hue="Bags", data=melted_df)
plt.show()
```

#### 20) Bin the data in the Total Volume column into four quantiles labeled ‘poor’, ‘modest’, ‘good’, and ‘excellent’, and store the bin labels in a new column. (point: 10)

```python
avocado_data['VolumeLevel'] = pd.qcut(avocado_data['TotalVolume'], q=4, labels=['poor', 'modest', 'good', 'excellent'])
avocado_data.head()
```
